## Checking for missing values

```python
penguin_data.isnull().sum()
```

we can see that there are two observations are having NULL values from these columns culmen_length_mm, culmen_depth_mm, flipper_length_mm and body_mass_g consists of 10 missing values.

```python
penguin_data.drop(penguin_data[penguin_data['body_mass_g'].isnull()].index, axis=0, inplace=True)
```

```python
penguin_data.drop(penguin_data[penguin_data['culmen_length_mm'].isnull()].index, axis=0, inplace=True)
```

```python
penguin_data.drop(penguin_data[penguin_data['culmen_depth_mm'].isnull()].index, axis=0, inplace=True)
```

```python
penguin_data.drop(penguin_data[penguin_data['flipper_length_mm'].isnull()].index, axis=0, inplace=True)
```

```python
penguin_data['sex'] = penguin_data['sex'].fillna('MALE')

#imputing the missing values in sex column with it's mode value
```

```python
penguin_data.isnull().sum()
```

```python
penguin_data.drop(penguin_data[penguin_data['sex']=='.'].index, inplace=True)
```

# Distribution among the variables

```python
print('CULMEN LENGTH DISTRIBUTION')
sns.violinplot(data=penguin_data,x="species",y="culmen_length_mm", size=8,inner="quartile")
plt.show()
```

```python
print("CULMEN DEPTH DISTRIBUTION")
sns.boxplot(data=penguin_data, x="species",hue="sex", y="culmen_depth_mm")
plt.show()
```

```python
print('Flipper Length Distribution')
sns.FacetGrid(penguin_data, hue="species", height=6,).map(sns.kdeplot, "flipper_length_mm",shade=True).add_legend()
plt.show()
```

```python
print('Body Mass Distribution')
sns.FacetGrid(penguin_data, hue="species", height=6,).map(sns.kdeplot, "body_mass_g",shade=True).add_legend()
plt.show()
```

## Extreme values of the variables

```python
print('Which island consists of most Penguins?')
print('Answer: Biscoe')
df = penguin_data['island'].value_counts().reset_index()

fig = sns.barplot(data=df, x='island', y='index')
fig.set(xlabel='', ylabel='ISLANDS')
plt.show
```

```python
print('Which species have highest culmen_length?')
print('Answer: Chinstrap(male and female)')
df = penguin_data.loc[:,['species','culmen_length_mm','sex']]
df['mean_culmen_length'] = df.groupby(['species','sex'])['culmen_length_mm'].transform('mean')
df = df.drop('culmen_length_mm', axis=1).drop_duplicates()

sns.barplot(data=df, x='mean_culmen_length', y='species', hue='sex')
plt.show()
```

```python
print('Which species have highest culmen_depth?')
print('Answer: Chinstrap(male and female)')
df = penguin_data.loc[:,['species','culmen_depth_mm','sex']]
df['mean_culmen_depth'] = df.groupby(['species','sex'])['culmen_depth_mm'].transform('mean')
df = df.drop('culmen_depth_mm', axis=1).drop_duplicates()

sns.barplot(data=df, x='mean_culmen_depth', y='species', hue='sex')
plt.show()
```

```python
print('Which species have highest flipper_length?')
print('Answer: Gentoo(male and female)')
df = penguin_data.loc[:,['species','flipper_length_mm','sex']]
df['mean_flipper_length'] = df.groupby(['species','sex'])['flipper_length_mm'].transform('mean')
df = df.drop('flipper_length_mm', axis=1).drop_duplicates()

sns.barplot(data=df, x='mean_flipper_length', y='species', hue='sex')
plt.show()
```

```python
print('Which species have highest body_mass?')
print('Answer: Gentoo(male and female) - Highly diverse values noticed')
df = penguin_data.loc[:,['species','body_mass_g','sex']]
df['mean_body_mass'] = df.groupby(['species','sex'])['body_mass_g'].transform('mean')
df = df.drop('body_mass_g', axis=1).drop_duplicates()

sns.barplot(data=df, x='mean_body_mass', y='species', hue='sex')
plt.show()
```
